```python
#imports
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
#eye detection
!wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_eye.xml
```

```python
# img1 will contain the image used for eye detection.
img1 = cv2.imread('/content/drive/MyDrive/Syeda subhana _ mini project/img1.png')
```

```python
eye_model=cv2.CascadeClassifier("/content/haarcascade_eye.xml")
eyes=eye_model.detectMultiScale(img1,1.34,5)
for (x,y,w,h) in eyes:
  cv2.rectangle(img1,(x,y),(x+w,y+h),(255,0,0),2)
cv2_imshow(img1)
```

```python
#face detection
!wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml
```

```python
#img2 will store the image used for face detection.
img2 = cv2.imread('/content/drive/MyDrive/Syeda subhana _ mini project/img2.webp')
```

```python
face_model = cv2.CascadeClassifier("/content/haarcascade_frontalface_default.xml")
faces = face_model.detectMultiScale(img2,1.2,5)
for (x,y,w,h) in faces:
  cv2.rectangle(img2,(x,y),(x+w,y+h),(0,255,0),4)
cv2_imshow(img2)
```
